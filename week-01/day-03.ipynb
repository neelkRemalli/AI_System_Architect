```python
import gradio as gr
from openai import OpenAI

# --- Configuration Layer ---
# Initialize the OpenAI client pointing to OpenRouter's API gateway
client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key="PASTE_YOUR_NEW_KEY_HERE", # Remember to use a fresh key!
)

def architect_playground(prompt, sys_msg, temp, tokens):
    """
    Core Logic: Handles the API call to the LLM.
    Uses system instructions and hyperparameters to control the output.
    """
    try:
        response = client.chat.completions.create(
            model="google/gemini-2.0-flash-001",
            messages=[
                {"role": "system", "content": sys_msg},
                {"role": "user", "content": prompt}
            ],
            temperature=temp,   # Controls randomness/creativity
            max_tokens=tokens    # Limits the length of the response
        )
        return response.choices[0].message.content
    except Exception as e:
        return f"Architectural Error: {str(e)}"

# --- UI Layer (Gradio Blocks) ---
with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown("# 🏢 AI System Architect - Lab 01")
    gr.Markdown("### Testing Model Personas and Inference Parameters")

    with gr.Row():
        with gr.Column():
            # Identity and prompt inputs
            sys_input = gr.Textbox(
                label="System Role (The Identity)",
                value="You are a senior system architect. Give technical and structured answers."
            )
            user_input = gr.Textbox(label="User Question", lines=4, placeholder="Ask something technical...")

            # Hyperparameter sliders
            t_slider = gr.Slider(0, 1.5, value=0.7, label="Temperature")
            tok_slider = gr.Slider(50, 1000, value=300, label="Max Tokens")

            run_btn = gr.Button("Execute Inference ⚡", variant="primary")

        # Output display area
        output = gr.Textbox(label="Model Output", lines=12, interactive=False)

    # Event handling: Maps inputs to the inference function
    run_btn.click(
        fn=architect_playground,
        inputs=[user_input, sys_input, t_slider, tok_slider],
        outputs=output
    )

# Launch the application
if __name__ == "__main__":
    demo.launch()
```
