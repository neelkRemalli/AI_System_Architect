```python
from transformers import pipeline
import torch
```

### Lab 1

```python
def prompt_processor(user_input, persona="helpful"):

    # Define the 'Role' of your bot
    if persona == "technical":
        role = "System: You are a Senior Network Engineer."
    else:
        role = "System: You are a Customer Support Assistant."

    # This is where you practice 'Prompt Engineering'
    template = f"{role}\nUser: {user_input}\nAssistant: Let's think step-by-step."
    
    # For now, we return the template to see how it looks
    return template

# Test the 'Chunked' logic
print(prompt_processor("My internet is slow.", persona="technical"))

   
```

```python
# Load the Modal
pipe = pipeline(
               "text-generation",
                model="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
                torch_dtype=torch.float16,)
```

```python
def chatbot_v1(user_query):
    messages = [
        {"role": "system", "content": "You are a professional support assistant."},
    ]
    
    prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    outputs = pipe(prompt, max_new_tokens=150, do_sample=True, temperature=0.7)

    
    return outputs[0]["generated_text"]


print(chatbot_v1("How do i fix a paper jam?"))
```

```python

pipe = pipeline("text-generation", 
                model="TinyLlama/TinyLlama-1.1B-Chat-v1.0", 
                torch_dtype=torch.bfloat16, 
                device_map="auto")

def chatbot_v1(user_query):
    # Chunk 2: Format the Prompt 
    # We use the 'Few-Shot' technique here
    messages = [
        {"role": "system", "content": "You are a professional support bot."},
        {"role": "user", "content": user_query},
    ]
    
    # Chunk 3: Generate Response 
    # This is the actual 'Inference' process
    prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    outputs = pipe(prompt, max_new_tokens=150, do_sample=True, temperature=0.7)
    
    return outputs[0]["generated_text"]

# Run the lab
print(chatbot_v1("How do I fix a paper jam?"))
```
